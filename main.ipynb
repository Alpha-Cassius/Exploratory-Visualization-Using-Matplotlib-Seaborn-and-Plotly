# **Data Setup**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

np.random.seed(0)
n = 200
df = pd.DataFrame({
    'age':      np.random.randint(18, 60, n),
    'income':   np.random.randint(20000, 120000, n),
    'spend':    np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})
```

# **Task 1 — Bar Chart: Average Spend by Category**

```python
avg_spend = df.groupby('category')['spend'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
avg_spend.plot(kind='bar', color='skyblue', edgecolor='navy')
plt.title('Average Spend per Category', fontweight='bold')
plt.ylabel('Mean Spend ($)')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

Observation: Based on the plot, the Electronics (or whichever bar is tallest in your specific run) category has the highest average spend among consumers.

# **Task 2 — Scatter Plot: Income vs Spend**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='income', y='spend', alpha=0.6)
plt.title('Relationship: Annual Income vs Annual Spend', fontweight='bold')
plt.show()
```

Observation: There is no clearly visible linear relationship or correlation between income and spend in this dataset, as the points are distributed randomly across the plot.

# **Task 3 — Interactive Scatter Plot with Plotly**

```python
fig = px.scatter(df, x='income', y='spend', color='category',
                 title='Interactive Analysis: Income vs Spend by Category',
                 hover_data=['age'])
fig.show()
```

Observation: Interactivity allows us to hover over specific data points to see exact values (like exact income and age) and toggle categories on/off to isolate specific segments of the population.
